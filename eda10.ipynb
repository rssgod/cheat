```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a population dataset
np.random.seed(0)  # for reproducibility
population = np.random.normal(loc=50, scale=10, size=10000)       #normal distribution

# Population statistics
pop_mean = np.mean(population)
pop_std = np.std(population)

print("Population Mean:", pop_mean)       # average
print("Population Std Dev:", pop_std)     # spread of data

# Step 2: Sampling process
sample_size = 50
num_samples = 200

sample_means = []           # store mean of each sample

# Step 3: Generate samples and compute sample means
for i in range(num_samples):
    sample = np.random.choice(population, size=sample_size)
    sample_mean = np.mean(sample)                              # mean of current sample
    sample_means.append(sample_mean)                           # Stores sample mean in list

# Step 4: Sampling distribution statistics
mean_of_sample_means = np.mean(sample_means)
std_of_sample_means = np.std(sample_means)                     # Standard deviation of sample means, Called Standard Error

print("\nMean of Sample Means:", mean_of_sample_means)
print("Std Dev of Sample Means:", std_of_sample_means)

# Step 5: Visualization
plt.figure(figsize=(10,5))
plt.hist(sample_means, bins=20, edgecolor='black')                         # bins=20 → number of bars
plt.axvline(pop_mean, linestyle='dashed', linewidth=2, label='Population Mean')          # Draws dashed line showing population mean
plt.axvline(mean_of_sample_means, linestyle='solid', linewidth=2, label='Sample Mean')
plt.title("Sampling Distribution of Sample Mean")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# Step 6: Demonstrating convergence
running_means = np.cumsum(sample_means) / np.arange(1, num_samples+1)           # np.cumsum() → cumulative sum of sample means, np.arange() → numbers from 1 to 200
# gives running average
plt.figure(figsize=(10,5))
plt.plot(running_means)
plt.axhline(pop_mean, linestyle='dashed', label='Population Mean')
plt.title("Convergence of Sample Mean")
plt.xlabel("Number of Samples")
plt.ylabel("Running Mean")
plt.legend()
plt.show()
```

A sampling distribution is the distribution of a statistic (like mean) obtained from multiple samples drawn from the same population. Sampling distribution is the distribution of sample statistics (like mean) obtained from repeated sampling.

Convergence is the process where the sample mean gets closer to the population mean as the number of samples increases.


Generates random samples from a population, computes their means, visualizes the sampling distribution, and demonstrates convergence of sample mean to population mean.
