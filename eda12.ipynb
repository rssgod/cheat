```python
import numpy as np

# Step 1: Create a population
population = np.arange(1, 101)

# Step 2: Parameters
sample_size = 10
num_samples = 30

# Step 3: Generate samples and calculate means
sample_means = []

for i in range(num_samples):
    sample = np.random.choice(population, sample_size)
    mean = float(np.mean(sample))   # convert to normal float
    sample_means.append(mean)
    print(f"Sample {i+1} Mean =", mean)

# Step 4: Final result
print("\nAll Sample Means:")
print(sample_means)
```
