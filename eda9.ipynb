```python
df['KM_Driven'].describe()
```

```python
sns.histplot(df['KM_Driven'],kde=True)
plt.tight_layout()
plt.show()
```

```python
df['Fuel'].value_counts()
```

```python
df['Fuel'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Fuel Type Count")
plt.xlabel("Fuel Type")
plt.ylabel("Count")
plt.show()
```

```python
plt.scatter(df['KM_Driven'],df['Selling_Price'])
plt.xlabel("KM Driven")
plt.ylabel("Selling Price")
plt.tight_layout()
plt.show()
```

```python
ct = pd.crosstab(df['Fuel'],df['Transmission'])
ct.plot(kind='bar')

plt.title('Fuel vs Transmission')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.show()
```

```python
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
